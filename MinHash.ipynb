Group the movies watched by each user

```python
user_movies = user_rating_df.groupby("userId")["movieId"].apply(list).to_dict()
```

Hash functions with three different methods: Division, Mid-Square and Folding

Util function: to make our max val for the hash function prime, based on the totalnumber of users

```python
def get_next_prime(num):
    while True:
        if num > 1:
            for i in range(2, int(num ** 0.5) + 1):
                if num % i == 0:
                    break
            else:
                return num  
        num += 1
```

```python
max_userid = len(user_movies)
p = get_next_prime(max_userid)
```

```python
# Division Hash Function
def division_hash(x):
    return x % p

# Mid-Square Hash Function
def mid_square_hash(x):
    squared = x * x
    squared_str = str(squared)
    mid_len = len(squared_str) // 2

    if len(squared_str) < 4:
        middle_digits = int(squared_str)  
    else:
        middle_digits = int(squared_str[mid_len-2:mid_len+2])  
    
    return middle_digits % p 

# Folding Hash Function
def folding_hash(x, num_parts=4):
    x_str = str(x)
    part_length = max(1, len(x_str) // num_parts)  
    sum_parts = 0

    # Loop through parts with valid slicing
    for i in range(0, len(x_str), part_length):
        part = x_str[i:i + part_length]  
        sum_parts += int(part)  

    return sum_parts % p  
```

Create MinHash Signature for each user by iterating through user's watched movie list and applying the hash functions

```python
def generate_hash_functions(num_hashes, max_val):
    np.random.seed(42)
    hash_funcs = []
    for _ in range(num_hashes):
        a = np.random.randint(1, max_val)
        b = np.random.randint(0, max_val)
        hash_funcs.append((a, b))
    return hash_funcs
```

For each user's movie list we compute the MinHash Signature based on the chosen hash method

```python
def generate_signature(user_movies, num_hashes, hash_method):
    max_val = max_userid
    hash_funcs = generate_hash_functions(num_hashes, max_val)
    signature = []
    
    for a, b in hash_funcs:
        min_hash = float('inf')
        for movie_id in user_movies:
            if hash_method == 'division':
                hash_value = division_hash(movie_id)
            elif hash_method == 'mid-square':
                hash_value = mid_square_hash(movie_id)
            elif hash_method == 'folding':
                hash_value = folding_hash(movie_id)
            min_hash = min(min_hash, hash_value)
        signature.append(min_hash)
    
    return signature
```

For each of the different hash methods we generate and store the Signature vectors for each user

```python
num_hashes = 50 
signatures = {}
hash_method = 'division'  

for user_id, movies in user_movies.items():
    signatures[user_id] = generate_signature(movies, num_hashes, hash_method)
```

Compare two signature vectors

```python
def jaccard_similarity(signature1, signature2):
    return np.mean(signature1 == signature2)
```

Find users with similarity to a target user that exceeds the threshold

```python
def find_similar_users(target_user_idx, signatures, threshold=0.8):
    similar_users = []
    target_signature = signatures[target_user_idx]
    
    for idx, signature in signatures.items():
        if idx != target_user_idx:  
            similarity = jaccard_similarity(target_signature, signature)
            if similarity >= threshold:
                similar_users.append((idx, similarity))
    
    similar_users.sort(key=lambda x: x[1], reverse=True)
    return similar_users
```

Change threshod to see the effect on grouping users

```python
# Define the range of thresholds to experiment with
thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]

target_user_idx = 1  

# Loop through each threshold and find similar users
for threshold in thresholds:
    similar_users = find_similar_users(target_user_idx, signatures, threshold)
    
    # Extract user IDs from similar_users
    user_ids = set([user_idx for user_idx, similarity in similar_users])
    
    print(f"\nThreshold: {threshold}")
    print(f"Users similar to user {target_user_idx} (Threshold: {threshold}): {user_ids}")
```

Iterate through the hash methods and the threshold values

```python
import time

# List of hash methods to test
hash_methods = ['division', 'mid-square', 'folding']
# List of thresholds to test
thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]

num_hashes = 50  # Number of hash functions
target_user_idx = 1  # Example target user

# Loop through each hash method
for hash_method in hash_methods:
    print(f"\n### Hash Method: {hash_method} ###")
    
    # Generate signatures using the current hash method
    signatures = {}
    start_time = time.time()
    for user_id, movies in user_movies.items():
        signatures[user_id] = generate_signature(movies, num_hashes, hash_method)
    generation_time = time.time() - start_time
    print(f"Signature generation time: {generation_time:.4f} seconds")
    
    # Loop through each threshold value
    for threshold in thresholds:
        start_time = time.time()
        similar_users = find_similar_users(target_user_idx, signatures, threshold)
        execution_time = time.time() - start_time
        
        user_ids = [user_idx for user_idx, similarity in similar_users]

        # Print results
        print(f"\nThreshold: {threshold}")
        print(f"Execution time: {execution_time:.4f} seconds")
        print(f"Users similar to user {target_user_idx}: {user_ids}")

```
